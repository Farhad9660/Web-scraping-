```python
import requests
import lxml
from bs4 import BeautifulSoup

# geting the web site first page (seed)
url = "https://www.metacritic.com/browse/movies/score/metascore/all/filtered?sort=desc"
headers = {
  'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/63.0.3239.132 Safari/537.36 QIHU 360SE'
}
r = requests.get(url, headers = headers)
print(r.status_code)

#saving the web page
path = "C:/Users/m.amini/Documents/Python Scripts/web_pages_meta/"
f = open(path + 'main.html', 'w')
f.write (r.text)
f.close()

# extract the urls from the page

soup = BeautifulSoup (r.content, "lxml")
frontier =[]
films = soup.find ('div', {'class': 'title_bump'}).find_all('a',{'class':'title'})
#print (films)

#iterate on links
counter = 0
for anc in films:
    film_url = "https://www.metacritic.com" + anc['href']
    frontier.append(film_url)
    
#print (frontier)

for link in frontier:
    print("film_link : " + link)
    r2 = requests.get(link, headers = headers)
    if r2.ok:
        counter +=1
        #f2 = open(path + str(counter) + ".html", 'w', encoding = 'utf-8')
        #f2.write(r2.text)
        #f2.close()
        
        soup2 = BeautifulSoup(r2.content, 'lxml')
        title = soup2.find('div', class_ = "product_page_title oswald").find('h1').text
        print("film title: " + title)
        #year = soup2.find('div', class_ = "product_page_title oswald").find('span').text
        year = soup2.select("div.product_page_title > span.release_year")[0].text 
        print("release_year: " + year)
        director = soup2.select("div.director > a > span")[0].text
        print("Director: " + director)
        genre = soup2.select("div.genres > span > span")
        print ("genre: ",  end ='')
        for e in genre:
            print(e.text + ", ", end ='')
        print ('')
        
        print ('----' *10)
```
