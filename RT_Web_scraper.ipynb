```python

import requests
import lxml
from bs4 import BeautifulSoup
import re

# gettin the web first page (seed)

url = "https://www.rottentomatoes.com/top/bestofrt/"

headers = {
  'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/63.0.3239.132 Safari/537.36 QIHU 360SE'
}
r = requests.get(url, headers = headers)
print (r.status_code)
# print (f.text)

# saving the web page
path = "C:/Users/m.amini/Documents/Python Scripts/web_pages/"
f = open (path +'main.html', 'w')
f.write(r.text)
f.close

# extract the urls from the page
soup = BeautifulSoup(r.content, 'lxml') 
frontier = []
films = soup.find ('table', {'class': 'table'}).find_all('a')
#print(films)

# iterate on the links
counter = 0
for anchor in films:
    film_url = 'https://www.rottentomatoes.com' + anchor['href']
    frontier.append(film_url)
    
for link in frontier:
    print(link)
    r2 = requests.get(link, headers = headers)
    if r2.ok:
        counter += 1 
        film_soup = BeautifulSoup(r2.content, 'lxml') 
        film_title = film_soup.find('h1', {'class':'scoreboard__title'})
        print ("film # " + str(counter) +': ' + film_title.string.strip())
        film_genre = film_soup.find('div', {'class':'meta-value genre'})
        film_genre = re.sub("\s+", " ", film_genre.string.strip())
        print ("Genre : " + film_genre)
        film_director = film_soup.find('a', {'data-qa':'movie-info-director'})
        print ("Director : " + film_director.string.strip())
        film_story = film_soup.find('div', {'class':'movie_synopsis clamp clamp-6 js-clamp'})
        print ("Story : \n" + film_story.string.strip())
        print('-------' * 10)
```
